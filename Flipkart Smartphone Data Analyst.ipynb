```python
brand_prices = flipkart.groupby('brand')[['original_price', 'discounted_price']].max()

brand_prices.plot(kind='bar', figsize=(15, 7))

# Customize the plot
plt.title('Average Original and Discounted Prices by Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.legend(['Original Price', 'Discounted Price'])

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# prompt: count plot for brand max to min nd showing count value

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 7))
sns.countplot(x='brand', data=flipkart, order=flipkart['brand'].value_counts().index)
plt.title('Count of Smartphones by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

# Annotate the bars with the count values
for p in plt.gca().patches:
    plt.gca().text(p.get_x() + p.get_width() / 2., p.get_height(), '%d' % int(p.get_height()),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# prompt: pie chart for top 10 brand

import matplotlib.pyplot as plt
# Calculate brand counts
brand_counts = flipkart['brand'].value_counts()

# Select top 10 brands
top_10_brands = brand_counts[:10]

# Create pie chart
plt.figure(figsize=(12,10))
plt.pie(top_10_brands, labels=top_10_brands.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Smartphone Brands')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
flipkart.groupby('battery_type').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Calculate the discount percentage
flipkart['discount_percentage'] = ((flipkart['original_price'] - flipkart['discounted_price']) / flipkart['original_price']) * 100

# Display the DataFrame with the new column
flipkart
```

```python
bins = [0, 10, 50, float('inf')]
labels = ['<10%', '10-50%', '>50%']

flipkart['discount_category'] = pd.cut(flipkart['discount_percentage'], bins=bins, labels=labels)

discount_category_counts = flipkart['discount_category'].value_counts()

plt.show()
```

```python
import plotly.express as px
import matplotlib.pyplot as plt

brand_discount = flipkart.groupby('brand')['discount_percentage'].mean().reset_index()

fig = px.scatter_3d(brand_discount, x='brand', y='discount_percentage', z='discount_percentage',
                   color='brand',
                   title='Brand vs. Discount Percentage (3D Scatter Plot)')
fig.show()


# 2D Bar Chart
plt.figure(figsize=(15, 7))
plt.bar(brand_discount['brand'], brand_discount['discount_percentage'])
plt.title('Average Discount Percentage by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: plg give me circle in both side with brand

import matplotlib.pyplot as plt
# Group by brand and calculate the count of items in each brand
brand_counts = flipkart['brand'].value_counts()

# Select the top 5 brands for better visualization in the pie chart
top_brands = brand_counts.head(5)

# Create a pie chart for the top brands
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.pie(top_brands, labels=top_brands.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(top_brands)))
plt.title('Top 5 Brands Distribution', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Group by brand and calculate the average discounted price
brand_avg_discounted_price = flipkart.groupby('brand')['discounted_price'].mean()

# Select the top 5 brands based on count for consistency
top_brands_avg_price = brand_avg_discounted_price[top_brands.index]

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.pie(top_brands_avg_price, labels=top_brands_avg_price.index, autopct='%1.2f', startangle=90, colors=sns.color_palette('magma', len(top_brands_avg_price)))
plt.title('Average Discounted Price by Top 5 Brands', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
plt.show()
```
